```python
import pandas as pd
import numpy as np
import time
import datetime
import calendar
```

## Historic

```python
df = pd.read_csv("BTCUSD_2018_1min_Raw.csv",index_col=False)
df.describe()
```

```python
df.head()
```

```python
pd.to_datetime(df.Date,errors='raise')
df.dtypes
```

```python
df['UnixTime'] = df.Date.apply(lambda x :int(calendar.timegm(datetime.datetime.strptime(x,"%d-%m-%y %H:%M").timetuple()))+ 259200)
df.head()
# (datetime.datetime.strptime('31-12-18 23:59',"%d-%m-%y %H:%M").timetuple())
```

```python
df.to_csv('BTCUSD_2018_1min_Outout.csv',sep=',',encoding='utf-8')
```

## Twitter

```python
df = pd.read_csv("mergefile.csv", index_col=False)
df.head()
```

```python
print(df.describe())
df.dropna(axis = 0,inplace=True)
print(df.describe())
df.head()
```

```python
df.dtypes
```

```python
df.timestamp =  pd.to_datetime(df.timestamp,errors='raise')
```

```python
df.head()
```

```python
df.dtypes
```

```python
# (df.timestamp - datetime.datetime(1970, 1, 1)).total_seconds() 
df['UnixTime'] = list(((date1 - datetime.datetime(1970, 1, 1)).total_seconds() for date1 in df.timestamp))
df.head()
```
