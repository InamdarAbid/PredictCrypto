```python
import pandas as pd
import numpy as np
import time
import datetime
import calendar
```

```python
df = pd.read_csv("BTCUSD_2018_1min.csv",index_col=False)
df.describe()
```

```python
df.head()
```

```python
pd.to_datetime(df.Date,errors='raise')
df.dtypes
```

```python
df['UnixTime'] = df.Date.apply(lambda x :int(calendar.timegm(datetime.datetime.strptime(x,"%d-%m-%y %H:%M").timetuple())) + 86400)
df.head()
# (datetime.datetime.strptime('31-12-18 23:59',"%d-%m-%y %H:%M").timetuple())
```

```python
df.to_csv('BTCUSD_2018_1min.csv',sep=',',encoding='utf-8')
```
