```python
def pre(row):
    row = re.sub(r'[^\x00-\x7F]+', '', str(row))
    return row

df['preprocess'] = df['preprocess'].apply(pre) 
    
```

Loading Pretrained Model

```python
with open("model_news_feeds", "rb") as f:
    mp = pickle.load(f)
```

```python
X_test = df.loc[:, "preprocess"]
X_test
```

Now first need to tokenize it

```python
tokenizer = Tokenizer(num_words=5000, split=" ")
tokenizer.fit_on_texts(X_test.values)

X = tokenizer.texts_to_sequences(X_test.values)
# X = pad_sequences(X, max_tokens=1814, padding="pre") # padding our text vector so they all have the same length
X[:10]
```

```python
X = pad_sequences(X,  max_tokens = None, padding="pre",  truncating="post", \
    value=0.0) # padding our text vector so they all have the same length
```

X = X[:, : 1814]

```python
predictions = mp.predict(X)
```

```python
print(predictions)
```

```python
df
```

```python
label = 0
l = []
count = 752
for i, prediction in enumerate(predictions):
    
    if np.argmax(prediction) == 0:
        label = -2
    elif np.argmax(prediction) == 1:
        label = -1
    elif np.argmax(prediction) == 2:
        label = 0
    elif np.argmax(prediction) == 3:
        label = 1
    else:
        label = 2
    l.append(label)
    df['Label (-2 to 2)'][count] = label
    count += 1
```
