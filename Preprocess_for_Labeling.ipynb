```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_14_Nov_2019_10Days.csv')
df.head()
```

```python
df.drop(columns = ['tweet-id','url','html'], inplace = True)
df.head()
```

```python
df = df[df.likes > 0]
df = df[df.replies > 0]
df = df[df.retweets > 0]
df.head()
df.reset_index(inplace = True, drop = True)
```
